```
# Day 3 Q1
def is_pangram(s):
    s = s.lower()
    for ch in 'abcdefghijklmnopqrstuvwxyz':
        if ch not in s:
            return False
    return True

print(is_pangram(input("Enter string: ")))
```

```
# Day 3 Q2
def longest_word(sentence):
    words = sentence.split()
    max_word = words[0]
    for w in words:
        if len(w) > len(max_word):
            max_word = w
    return max_word

print(longest_word(input("Enter sentence: ")))
```

```
# Day 3 Q3
def count_case(s):
    upper = 0
    lower = 0
    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
    return upper, lower

u, l = count_case(input("Enter string: "))
print("Uppercase:", u, "Lowercase:", l)
```

```
# Day 7 Q3
students = {}
n = int(input("Enter number of students: "))
for i in range(n):
    name = input("Name: ")
    marks = int(input("Marks: "))
    students[name] = marks
max_marks = max(students.values())
for name, marks in students.items():
    if marks == max_marks:
        print(name)
```
